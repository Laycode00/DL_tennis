```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
file_path = '4_all.csv'
df = pd.read_csv(file_path)
df.head(10)
```

```python
df.info()
```

```python
# 'binary' 열에서 1이 연속으로 3번 나오는 인덱스 찾기
consecutive_ones = df['binary'].rolling(window=3).apply(lambda x: all(x == 1), raw=True).dropna().astype(bool)
indices = list(consecutive_ones[consecutive_ones].index)

# 인덱스 랜덤 셔플
np.random.shuffle(indices)

# 앞뒤로 300개 행을 포함하여 자르기
sliced_dfs = []

for idx in indices:
    # 앞뒤로 300개 행을 포함한 범위 계산
    start = max(0, idx - 149)
    end = min(len(df), idx + 149)  # +301로 수정하여 1을 포함한 범위를 유지
    
    # 해당 범위의 행을 서브셋에 추가
    sliced_dfs.append(df.iloc[start:end])

# 서브셋을 결합하여 최종 데이터프레임 생성
final_df = pd.concat(sliced_dfs).reset_index(drop=True)

# 결과 확인
print(final_df)

# 최종 데이터프레임을 CSV 파일로 저장
final_df.to_csv('sliced_data.csv', index=False)
```

```python
column_name = 'binary'
zero_count = (df[column_name] == 0).sum()
one_count = (df[column_name] == 1).sum()

print(f"'{column_name}' 컬럼에서 0의 개수: {zero_count}")
print(f"'{column_name}' 컬럼에서 1의 개수: {one_count}")
```
